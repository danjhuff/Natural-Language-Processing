```python
import pandas as pd
import re

def extract_votes(row):
    text = row['Text']
    file_name = row['File_Name']
    votes = []
    
    # Find all AYES and NOES section positions
    ayes_matches = list(re.finditer(r'AYES:\s*(.*?)NOES:', text, re.DOTALL))
    noes_matches = list(re.finditer(r'NOES:\s*(.*?)(ABSTENTIONS:|The "Ayes" have it|Hon, Senato|Next order)', text, re.DOTALL))
    
    # Make sure pairs align (use min length of matches)
    num_pairs = min(len(ayes_matches), len(noes_matches))
    
    for i in range(num_pairs):
        ayes_text = ayes_matches[i].group(1)
        noes_text = noes_matches[i].group(1)
        question_number = i + 1  # Question index starting from 1
        
        # Match "Sen. Name", "Hon. Name", and "Sen (Dr.) Name"
        ayes_names = re.findall(r'(?:Sen(?:\s*\(Dr\.\))?|Hon)\.\s*([^;,\n]+)', ayes_text)
        noes_names = re.findall(r'(?:Sen(?:\s*\(Dr\.\))?|Hon)\.\s*([^;,\n]+)', noes_text)
        
        for name in ayes_names:
            votes.append({
                'Name': name.strip(),
                'Vote': 'Aye',
                'File_Name': file_name,
                'Question': question_number
            })
            
        for name in noes_names:
            votes.append({
                'Name': name.strip(),
                'Vote': 'No',
                'File_Name': file_name,
                'Question': question_number
            })
            
    return votes

# Apply to all rows
all_votes = []
for _, row in voting_df.iterrows():
    all_votes.extend(extract_votes(row))

# Create the final DataFrame
vote_df = pd.DataFrame(all_votes)

vote_df
```

```python
import pandas as pd
import re

# Function to extract the senator name from the Speaker column
def extract_senator_name(speaker_text):
    """
    Extract senator name from speaker text, handling different title formats:
    - Sen. Name
    - Hon. Name
    - Sen (Dr.) Name
    """
    if pd.isna(speaker_text):
        return None
    
    pattern = r'(?:Sen(?:\s*\(Dr\.\))?|Hon)\.?\s+([A-Za-z\s]+)'
    match = re.search(pattern, speaker_text)
    
    if match:
        return match.group(1).strip()  # Return just the name part, trimmed
    
    # If no title match is found but there's text, check if it might be just a name
    if isinstance(speaker_text, str) and speaker_text.strip():
        # Check if the text contains what looks like a proper name (capitalized word)
        name_match = re.search(r'([A-Z][a-z]+(?:\s+[A-Z][a-z]+)*)', speaker_text)
        if name_match:
            return name_match.group(1).strip()
    
    return None

# Extract senator names from the Speaker column
filtered_df['Senator_Name'] = filtered_df['Speaker'].apply(extract_senator_name)

# Create a dictionary mapping (senator_name, file_name) to their votes
# This ensures we match votes from the same file
vote_dict = {}
for _, row in vote_df.iterrows():
    key = (row['Name'], row['File_Name'])
    vote_dict[key] = row['Vote']

# Function to look up the vote for each senator, ensuring file_name match
def get_vote(row):
    if pd.isna(row['Senator_Name']):
        return None
    
    # Try to get vote using both name and file_name
    key = (row['Senator_Name'], row['File_Name'])
    vote = vote_dict.get(key)
    
    # If no vote found for this specific file, report as "No vote recorded"
    if vote is None:
        return "No vote recorded"
    
    return vote

# Add the Vote column to filtered_df
filtered_df['Vote'] = filtered_df.apply(get_vote, axis=1)
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.metrics import cohen_kappa_score, make_scorer, f1_score
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.feature_selection import SelectFromModel
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Download required NLTK resources
try:
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    nltk.download('wordnet', quiet=True)
except:
    print("NLTK resources might need to be downloaded manually")

# Using your existing dataframe
training_df = filtered_with_votes.copy()
training_df = training_df[['Text', 'Vote']]

print("Original DataFrame:")
print(training_df.head())
print("\nShape:", training_df.shape)
print("\nColumns:", training_df.columns.tolist())

# Check the distribution of votes
print("\nVote distribution:")
vote_counts = training_df['Vote'].value_counts()
print(vote_counts)

# Visualize class distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='Vote', data=training_df)
plt.title('Class Distribution')
plt.tight_layout()
plt.show()

# Calculate class weights for handling imbalance
class_weights = {class_: len(training_df) / (len(vote_counts) * count) 
                for class_, count in vote_counts.items()}
print("\nClass weights for handling imbalance:")
print(class_weights)

# 1. Text preprocessing function (unchanged)
def preprocess_text(text):
    """Clean and preprocess text data"""
    if not isinstance(text, str):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove special characters and digits
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Tokenize text
    tokens = word_tokenize(text)
    
    # Remove stopwords - using a smaller set of stopwords for stance detection
    stop_words = set(stopwords.words('english'))
    # Remove some stopwords that might be important for stance detection
    stance_relevant = {'not', 'no', 'nor', 'against', 'for', 'but', 'however', 'although', 
                      'though', 'despite', 'while', 'yet', 'still'}
    filtered_stop_words = stop_words - stance_relevant
    tokens = [word for word in tokens if word not in filtered_stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    # Join tokens back into text
    cleaned_text = ' '.join(tokens)
    
    return cleaned_text

# Apply preprocessing to the text column
training_df['processed_text'] = training_df['Text'].apply(preprocess_text)

print("\nPreprocessed Text Example:")
print(training_df['processed_text'].head())

# 2. Add Basic Text Features
training_df['text_length'] = training_df['Text'].apply(len)
training_df['word_count'] = training_df['Text'].apply(lambda x: len(str(x).split()))
training_df['avg_word_length'] = training_df['Text'].apply(lambda x: 
                                       np.mean([len(word) for word in str(x).split()]) 
                                       if len(str(x).split()) > 0 else 0)

# 3. Feature Engineering
# Extract features from text using TF-IDF with improved parameters
tfidf_vectorizer = TfidfVectorizer(
    max_features=1500,
    ngram_range=(1, 3),
    min_df=3,
    max_df=0.9
)
X_tfidf = tfidf_vectorizer.fit_transform(training_df['processed_text'])

print("\nTF-IDF Features Shape:", X_tfidf.shape)

# Create feature names for later analysis
feature_names = tfidf_vectorizer.get_feature_names_out()

# Prepare the target variable
y = training_df['Vote']

# Create additional numerical features
additional_features = training_df[['text_length', 'word_count', 'avg_word_length']].values

# Combine TF-IDF features with numerical features
from scipy.sparse import hstack
X_combined = hstack([X_tfidf, additional_features])

print("\nCombined Features Shape:", X_combined.shape)

# 4. Split data with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.3, random_state=42, stratify=y
)

print("\nTraining Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)

# 5. Model Comparison - using Cohen's Kappa as primary metric
# Creating both F1 and Kappa scorers for comparison
f1_scorer = make_scorer(f1_score, average='weighted')
kappa_scorer = make_scorer(cohen_kappa_score)  # Added kappa scorer

classifiers = {
    'RandomForest': RandomForestClassifier(random_state=42, class_weight='balanced'),
    'LogisticRegression': LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000),
}

results = {}

for name, classifier in classifiers.items():
    print(f"\nTraining {name}...")
    
    # Create pipeline with feature selection and SMOTE for class imbalance
    pipeline = ImbPipeline([
        ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42))),
        ('sampling', SMOTE(random_state=42)),
        ('classifier', classifier)
    ])
    
    # Cross-validation scores using kappa
    cv_kappa_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring=kappa_scorer)
    cv_kappa = np.mean(cv_kappa_scores)
    print(f"{name} 5-fold CV Kappa Score: {cv_kappa:.4f}")
    
    # Also calculate F1 for comparison
    cv_f1_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring=f1_scorer)
    cv_f1 = np.mean(cv_f1_scores)
    print(f"{name} 5-fold CV F1 Score: {cv_f1:.4f}")
    
    # Train model on full training set
    pipeline.fit(X_train, y_train)
    
    # Predict
    y_pred = pipeline.predict(X_test)
    
    # Evaluate - now using Kappa as primary metric
    accuracy = accuracy_score(y_test, y_pred)
    kappa = cohen_kappa_score(y_test, y_pred)  # Calculate Cohen's Kappa
    f1 = f1_score(y_test, y_pred, average='weighted')
    report = classification_report(y_test, y_pred, output_dict=True)
    
    # Store results
    results[name] = {
        'accuracy': accuracy,
        'kappa_score': kappa,
        'f1_score': f1,
        'cv_kappa_score': cv_kappa,
        'cv_f1_score': cv_f1,
        'report': report,
        'model': pipeline
    }
    
    print(f"{name} Test Kappa Score: {kappa:.4f}")
    print(f"{name} Test F1 Score: {f1:.4f}")
    print(f"{name} Test Accuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))
    
    # Plot Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=np.unique(y), 
                yticklabels=np.unique(y))
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - {name}')
    plt.show()

# Plot comparison of metrics across models
models = list(results.keys())
kappa_scores = [results[model]['kappa_score'] for model in models]
f1_scores = [results[model]['f1_score'] for model in models]
accuracy_scores = [results[model]['accuracy'] for model in models]

# Create grouped bar chart for comparison
plt.figure(figsize=(12, 6))
x = np.arange(len(models))
width = 0.25

# Create bars with values on top
bars1 = plt.bar(x - width, kappa_scores, width, label='Kappa')
bars2 = plt.bar(x, f1_scores, width, label='F1')
bars3 = plt.bar(x + width, accuracy_scores, width, label='Accuracy')

# Add values on top of the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'{height:.3f}', ha='center', va='bottom', fontsize=9)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Model Performance Comparison')
plt.xticks(x, models)
plt.ylim(0, 1.05)
plt.legend()
plt.tight_layout()
plt.show()

# Find best model based on Kappa score
best_model_name = max(results, key=lambda x: results[x]['kappa_score'])
best_model = results[best_model_name]['model']
print(f"\nBest Model: {best_model_name} with Kappa Score {results[best_model_name]['kappa_score']:.4f}")

# 6. Error Analysis for the best model
y_pred_best = best_model.predict(X_test)
y_proba_best = best_model.predict_proba(X_test)

# Identify misclassified examples
misclassified_mask = y_pred_best != y_test.values
misclassified_indices = np.where(misclassified_mask)[0]

# Safe way to create misclassified examples dataframe
# Get the indices in the original dataframe
test_indices = y_test.index[misclassified_indices].tolist()

# Create the misclassified examples dataframe
misclassified_df = pd.DataFrame({
    'Text': training_df.loc[test_indices, 'Text'].values,
    'Processed_Text': training_df.loc[test_indices, 'processed_text'].values,
    'Actual_Vote': y_test.iloc[misclassified_indices].values,
    'Predicted_Vote': y_pred_best[misclassified_indices],
    'Confidence': [np.max(prob) for prob in y_proba_best[misclassified_indices]]
})

print("\nSample of Misclassified Examples:")
print(misclassified_df.head())
```
